```python
import numpy as np
from netCDF4 import Dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ddir = '/home/robbie/Dropbox/kuka_mosaic/runway_nc/'
d_ka = pd.read_csv(f'{ddir}kuka_stare_decon_Ka-Scat20200119-111535.csv')
d_ku = pd.read_csv(f'{ddir}kuka_stare_decon_Ku-Scat20200119-111535.csv')

for d in [d_ka,d_ku]:
    x=np.diff(d['time'])
    # plt.hist(x,bins=np.arange(0.3,0.6,0.002))
    # plt.show()
    mean=np.nanmean(x)
    print(np.round(mean,decimals=3),
          np.round(1/mean,decimals=3))
```
